# Regress√£o Linear Manual ‚Äî Estudo Intuitivo e Visual


### Ideia Intuitiva da Regress√£o Linear

Imagine que voc√™ tem uma nuvem de pontos num gr√°fico ‚Äî cada ponto representa uma observa√ß√£o real de algum fen√¥meno. Por exemplo, voc√™ pode estar observando como o n√∫mero de horas estudadas (no eixo X) afeta a nota na prova (no eixo Y).

A ideia da regress√£o linear √© encontrar uma **linha reta** que passe "o mais pr√≥ximo poss√≠vel" desses pontos. Essa linha √© chamada de **reta de regress√£o**. Ela nos permite:

- Entender a tend√™ncia geral dos dados (por exemplo, mais estudo leva a notas maiores?)
- Prever novos valores (se algu√©m estudar 5 horas, qual seria a nota esperada?)

### Por que uma linha?
Porque estamos assumindo que a rela√ß√£o entre as vari√°veis √© **linear**, ou seja, segue o padr√£o:

\[ y = a \cdot x + b \]

Onde:
- **x** √© a vari√°vel explicativa (ex: horas de estudo)
- **y** √© a vari√°vel dependente (ex: nota)
- **a** √© o coeficiente angular (inclina√ß√£o da linha)
- **b** √© o coeficiente linear (ponto onde a linha cruza o eixo y)

### O que significa a inclina√ß√£o?
A inclina√ß√£o (coeficiente **a**) indica o quanto **y** muda quando **x** aumenta uma unidade:
- Se **a > 0**, a linha sobe (rela√ß√£o positiva)
- Se **a < 0**, a linha desce (rela√ß√£o negativa)
- Se **a = 0**, a linha √© horizontal (sem rela√ß√£o entre as vari√°veis)

### Como escolher a melhor linha?
A "melhor" linha √© a que minimiza a soma dos quadrados das dist√¢ncias verticais entre os pontos reais e a linha. Essa t√©cnica √© chamada de **m√≠nimos quadrados**.

√â como se tent√°ssemos ajustar uma r√©gua sobre os dados de forma que ela fique o mais equilibrada poss√≠vel entre todos os pontos.

### Aplica√ß√µes
- Economia: prever pre√ßos com base na demanda
- Medicina: relacionar idade com press√£o arterial
- Engenharia: prever desgaste de pe√ßas com base no tempo de uso
- Machine Learning: √© a base para algoritmos mais avan√ßados

### üßæ Conclus√£o: O que aprendemos com a Regress√£o Linear Manual

Ao longo desta se√ß√£o, exploramos a **regress√£o linear** de forma pr√°tica e intuitiva, partindo de uma simples nuvem de pontos at√© a constru√ß√£o manual de uma reta preditiva. Entendemos:

1. **O que √© regress√£o linear** e por que usamos uma reta para representar a rela√ß√£o entre duas vari√°veis.
2. **Como calcular manualmente os coeficientes** `a` (inclina√ß√£o) e `b` (intercepto) da reta.
3. **Como prever valores** com base na equa√ß√£o da reta.
4. **Como avaliar a qualidade da reta**, analisando os res√≠duos (erros) e calculando o **Erro Quadr√°tico M√©dio (MSE)**.
5. Como **visualizar o modelo**, tanto a reta quanto os res√≠duos.

Essa abordagem **refor√ßa o entendimento matem√°tico e visual** por tr√°s dos modelos de regress√£o ‚Äî antes de utilizar bibliotecas prontas.

### C√°lculo Manual dos Coeficientes da Regress√£o Linear

Vamos aprender como calcular a reta de regress√£o linear na pr√°tica, **sem usar bibliotecas autom√°ticas**.

Dada a equa√ß√£o da reta:
\[
y = a \cdot x + b
\]

Os coeficientes `a` (inclina√ß√£o) e `b` (intercepto) podem ser calculados manualmente pelas f√≥rmulas:

\[
a = \frac{n \sum(xy) - \sum x \sum y}{n \sum(x^2) - (\sum x)^2}
\]
\[
b = \frac{\sum y - a \sum x}{n}
\]

Onde:
- \( n \) √© o n√∫mero de pontos
- \( \sum x \), \( \sum y \), \( \sum(xy) \), \( \sum(x^2) \) s√£o as somas dos dados

Vamos implementar isso passo a passo em Python usando um exemplo simples.


```python
# Importa√ß√£o das Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

### üî¢ Dados simulados

```python
x = np.array([1.5, 1.6, 1.7, 1.8, 1.9])  # Altura (em metros)
y = np.array([45, 55, 65, 72, 85])       # Peso (em kg)
```

### üßÆ C√°lculo dos coeficientes da reta

```python
n = len(x)
soma_x = np.sum(x)
soma_y = np.sum(y)
soma_xy = np.sum(x * y)
soma_x2 = np.sum(x ** 2)
```

### F√≥rmulas da regress√£o linear simples

```python
m = (n * soma_xy - soma_x * soma_y) / (n * soma_x2 - soma_x ** 2)
b = (soma_y - m * soma_x) / n

print(f"Equa√ß√£o da reta: y = {m:.2f}x + {b:.2f}")
```

### üìà Visualiza√ß√£o
### Gerar valores previstos

```python
y_pred = m * x + b
```

### Gr√°fico de dispers√£o com a reta

```python
plt.scatter(x, y, color='blue', label='Dados reais')
plt.plot(x, y_pred, color='red', label='Reta ajustada')
plt.xlabel('Altura (m)')
plt.ylabel('Peso (kg)')
plt.title('Regress√£o Linear Manual')
plt.legend()
plt.grid(True)
plt.show()
```

### C√°lculo Manual dos Coeficientes da Regress√£o Linear

Vamos aprender como calcular a reta de regress√£o linear na pr√°tica, **sem usar bibliotecas autom√°ticas**.

Dada a equa√ß√£o da reta:
\[
y = a \cdot x + b
\]

Os coeficientes `a` (inclina√ß√£o) e `b` (intercepto) podem ser calculados manualmente pelas f√≥rmulas:

\[
a = \frac{n \sum(xy) - \sum x \sum y}{n \sum(x^2) - (\sum x)^2}
\]
\[
b = \frac{\sum y - a \sum x}{n}
\]

Onde:
- \( n \) √© o n√∫mero de pontos
- \( \sum x \), \( \sum y \), \( \sum(xy) \), \( \sum(x^2) \) s√£o as somas dos dados

Vamos implementar isso passo a passo em Python usando um exemplo simples.


```python
# Dados de exemplo: horas de estudo (x) e nota na prova (y)
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

# N√∫mero de pontos
n = len(x)

# C√°lculos intermedi√°rios
soma_x = sum(x)
soma_y = sum(y)
soma_xy = sum([xi * yi for xi, yi in zip(x, y)])
soma_x2 = sum([xi**2 for xi in x])

# C√°lculo da inclina√ß√£o (a) e do intercepto (b)
a = (n * soma_xy - soma_x * soma_y) / (n * soma_x2 - soma_x**2)
b = (soma_y - a * soma_x) / n

print(f"Inclina√ß√£o (a): {a:.2f}")
print(f"Intercepto (b): {b:.2f}")
```

```python
# Pontos observados
plt.scatter(x, y, color='blue', label='Pontos observados')

# Reta de regress√£o calculada
x_vals = list(range(min(x), max(x) + 1))
y_vals = [a * xi + b for xi in x_vals]
plt.plot(x_vals, y_vals, color='red', label='Reta de regress√£o')

# Configura√ß√µes do gr√°fico
plt.xlabel("Horas de Estudo")
plt.ylabel("Nota na Prova")
plt.title("Regress√£o Linear Manual")
plt.legend()
plt.grid(True)
plt.show()
```

### Previs√µes com a Reta de Regress√£o

Depois de ajustar a reta de regress√£o, podemos us√°-la para prever valores novos.

Se a equa√ß√£o da reta for:

\[
y = a \cdot x + b
\]

Basta substituir o valor de **x** (ex: horas de estudo) para prever **y** (ex: nota esperada).

Esse √© o grande poder da regress√£o: usar padr√µes do passado para prever o futuro.


### üßÆ Prever manualmente a nota com base nas horas de estudo

```python
def prever_nota(horas_estudo, a, b):
    return a * horas_estudo + b

# Coeficientes calculados manualmente
a = 2.5
b = 3.0

# Exemplo de previs√µes
horas = [2, 4, 6, 8]
for h in horas:
    print(f"Se estudar {h} horas, a nota prevista √©: {prever_nota(h, a, b):.2f}")
```

### üî¢ Dados simulados

```python
horas_estudo = [1, 2, 3, 4, 5]
notas = [4, 6, 7, 10, 11]
```

### üßÆ C√°lculo dos coeficientes da reta

```python
# Dados
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 6, 7, 10, 11])
n = len(x)

# Somat√≥rios necess√°rios
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x_squared = np.sum(x**2)

# C√°lculo do coeficiente angular (a)
a = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)

# C√°lculo do coeficiente linear (b)
b = (sum_y - a * sum_x) / n

print(f"Inclina√ß√£o (a): {a:.2f}")
print(f"Intercepto (b): {b:.2f}")
```

### üìä Plotando os pontos e a reta

```python
# Pontos reais
plt.scatter(x, y, label="Dados reais", color="blue")

# Reta de regress√£o
x_vals = np.array([min(x), max(x)])
y_vals = a * x_vals + b
plt.plot(x_vals, y_vals, label="Reta de regress√£o", color="red")

# Configura√ß√µes do gr√°fico
plt.xlabel("Horas de Estudo")
plt.ylabel("Nota")
plt.title("Regress√£o Linear Manual")
plt.legend()
plt.grid(True)
plt.show()
```

### üìâ Res√≠duos: o erro entre previsto e real

#### O que s√£o res√≠duos?

O **res√≠duo** √© a diferen√ßa entre o valor real (**y**) e o valor previsto pela regress√£o (**≈∑**):

\[
\text{res√≠duo} = y - \hat{y}
\]

Cada ponto tem um erro associado. Se a previs√£o estiver perfeita, o res√≠duo ser√° 0. Mas isso raramente acontece.

#### C√°lculo dos res√≠duos no exemplo

```python
# Valor previsto pela regress√£o para cada x
y_pred = a * x + b

# Res√≠duos (diferen√ßa entre real e previsto)
residuos = y - y_pred

# Exibir os res√≠duos
for i in range(n):
    print(f"x={x[i]:.0f}, y real={y[i]:.1f}, y previsto={y_pred[i]:.2f}, res√≠duo={residuos[i]:+.2f}")
```

### üéØ Como saber se a regress√£o foi boa ?

```python
# Soma dos quadrados dos res√≠duos
SSR = np.sum(residuos**2)
print(f"Soma dos quadrados dos res√≠duos (SSR): {SSR:.2f}")
```

### üìä Visualizando os res√≠duos no gr√°fico

Uma boa forma de **entender onde a regress√£o erra mais ou menos** √© desenhar as linhas dos res√≠duos (do ponto real at√© o ponto previsto na reta).

```python
# Gr√°fico com pontos reais e reta de regress√£o
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Dados reais')
plt.plot(x, y_pred, color='red', label='Reta de regress√£o')

# Adiciona as linhas verticais dos res√≠duos
for i in range(n):
    plt.plot([x[i], x[i]], [y[i], y_pred[i]], color='gray', linestyle='--')

# Personaliza√ß√£o
plt.title("Visualiza√ß√£o dos res√≠duos")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

### üìä Gr√°fico de Previs√µes com a reta de regress√£o

```python
# Novos valores de x (fora do conjunto original)
x_novos = np.array([1.5, 2.5, 3.5, 5.5, 6.5])

# Previs√£o dos valores de y usando a equa√ß√£o da reta
y_novos = a * x_novos + b

# Gr√°fico com dados reais, reta e previs√µes novas
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Dados reais')
plt.plot(x, y_pred, color='red', label='Reta de regress√£o')
plt.scatter(x_novos, y_novos, color='green', label='Novas previs√µes', marker='x', s=100)

plt.title("Previs√µes com a reta de regress√£o")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

### ‚úÖ Interpreta√ß√£o
Os pontos verdes representam as previs√µes feitas com a reta para valores novos de x.

Isso √© exatamente o que modelos de Machine Learning fazem: usam um padr√£o aprendido para prever novos dados.

### M√©trica R¬≤ ‚Äî O quanto a reta explica os dados?

Depois de ajustar uma reta aos dados, queremos saber: **o qu√£o boa ela √©?**

A m√©trica mais comum para isso √© o **R¬≤ (coeficiente de determina√ß√£o)**.

---

#### Intui√ß√£o do R¬≤:

Imagine que voc√™ tem os seguintes dois cen√°rios:

1. **Sem reta nenhuma**: A √∫nica previs√£o que voc√™ pode fazer √© a **m√©dia dos valores de y**.
2. **Com a reta ajustada**: Agora voc√™ tem previs√µes mais pr√≥ximas dos valores reais, porque considerou a rela√ß√£o entre x e y.

O **R¬≤ compara o quanto de varia√ß√£o dos dados √© explicada pela reta** em rela√ß√£o ao que seria explicado apenas usando a m√©dia.

---

#### F√≥rmula do R¬≤:

\[
R^2 = 1 - \frac{S_{res}}{S_{tot}}
\]

Onde:
- \( S_{res} \) = Soma dos res√≠duos ao quadrado (erros entre os valores previstos e os reais)
- \( S_{tot} \) = Soma total dos quadrados (erros entre os valores reais e a m√©dia de y)

---

#### Interpreta√ß√£o:

- **R¬≤ = 1**: A reta explica 100% da varia√ß√£o dos dados (perfeita!)
- **R¬≤ = 0**: A reta n√£o explica nada, t√£o ruim quanto usar s√≥ a m√©dia.
- **Entre 0 e 1**: Quanto mais perto de 1, melhor a reta explica os dados.

---

#### Exemplo visual:
Se os pontos estiverem bem alinhados com a reta, o R¬≤ ser√° alto.  
Se os pontos estiverem espalhados, o R¬≤ ser√° baixo.

---

#### Observa√ß√£o:
R¬≤ √© √∫til para avaliar **modelos lineares simples**, mas pode enganar em modelos mais complexos ou quando h√° outliers. Por isso, **visualizar os res√≠duos** e **entender o contexto dos dados** tamb√©m √© fundamental.

---

### üìå C√≥digo: Visualizando a Reta e o R¬≤

```python
# Dados fict√≠cios
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Ajuste manual dos coeficientes (ou use np.polyfit para testar)
a = 0.7
b = 2
y_pred = a * x + b

# C√°lculo do R¬≤
y_mean = np.mean(y)
ss_total = np.sum((y - y_mean) ** 2)
ss_res = np.sum((y - y_pred) ** 2)
r2 = 1 - (ss_res / ss_total)

# Visualiza√ß√£o
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Dados reais', color='blue')
plt.plot(x, y_pred, label='Reta ajustada', color='red')
plt.title(f'Regress√£o Linear - R¬≤ = {r2:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

### üìä Compara√ß√£o Visual: R¬≤ Alto vs. R¬≤ Baixo

```python
# Caso 1: Dados bem ajustados (R¬≤ alto)
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
a1, b1 = 2, 0
y1_pred = a1 * x1 + b1
ss_total1 = np.sum((y1 - np.mean(y1)) ** 2)
ss_res1 = np.sum((y1 - y1_pred) ** 2)
r2_1 = 1 - ss_res1 / ss_total1

# Caso 2: Dados com dispers√£o (R¬≤ baixo)
x2 = np.array([1, 2, 3, 4, 5])
y2 = np.array([2, 5, 3, 7, 4])
a2, b2 = 1, 2
y2_pred = a2 * x2 + b2
ss_total2 = np.sum((y2 - np.mean(y2)) ** 2)
ss_res2 = np.sum((y2 - y2_pred) ** 2)
r2_2 = 1 - ss_res2 / ss_total2

# Plotagem lado a lado
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

ax1.scatter(x1, y1, color='blue', label='Dados')
ax1.plot(x1, y1_pred, color='red', label='Reta ajustada')
ax1.set_title(f'R¬≤ Alto = {r2_1:.2f}')
ax1.legend()
ax1.grid(True)

ax2.scatter(x2, y2, color='blue', label='Dados')
ax2.plot(x2, y2_pred, color='red', label='Reta ajustada')
ax2.set_title(f'R¬≤ Baixo = {r2_2:.2f}')
ax2.legend()
ax2.grid(True)

plt.suptitle('Comparando R¬≤: Ajuste Bom vs. Ajuste Ruim', fontsize=16)
plt.show()
```

### üßÆ Exemplo com C√°lculo Manual

```python
# Dados de exemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# M√©dias
x_mean = np.mean(x)
y_mean = np.mean(y)

# C√°lculo da covari√¢ncia (manual)
cov_xy = np.mean((x - x_mean) * (y - y_mean))

# C√°lculo da vari√¢ncia de x
var_x = np.mean((x - x_mean)**2)

# Coeficientes da regress√£o
a = cov_xy / var_x
b = y_mean - a * x_mean

# Resultado
print(f"Inclina√ß√£o (a): {a:.2f}")
print(f"Intercepto (b): {b:.2f}")
print(f"Equa√ß√£o da reta: y = {a:.2f}x + {b:.2f}")
```

### üìâ Visualiza√ß√£o da Reta com os Dados

```python
# Predi√ß√µes com a reta
y_pred = a * x + b

# Plotando
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Dados Reais')
plt.plot(x, y_pred, color='red', label='Reta de Regress√£o')
plt.title("Reta de Regress√£o Calculada Manualmente")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

### üìä Explica√ß√£o Visual: O que Significa a Reta Ajustada?
A reta de regress√£o representa a tend√™ncia central da rela√ß√£o entre as vari√°veis x e y. Para cada valor de x, ela estima um valor de y, chamado de valor previsto ou ≈∑ (y chap√©u).

Podemos visualizar isso destacando:

Os pontos reais (dados observados)

Os pontos previstos (que est√£o sobre a reta)

As setas dos res√≠duos: linhas que conectam os valores reais aos previstos (indicando o erro de previs√£o)


### üìê Visualizando os Res√≠duos

```python
# C√°lculo dos res√≠duos
residuos = y - y_pred

# Gr√°fico com res√≠duos
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Dados Reais')
plt.plot(x, y_pred, color='red', label='Reta de Regress√£o')

# Linhas verticais dos res√≠duos
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], y_pred[i]], 'k--', linewidth=1)

# Anota√ß√µes e estilo
plt.title("Visualiza√ß√£o dos Res√≠duos")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

Se houver um padr√£o nos res√≠duos (curvas, tend√™ncia), talvez a rela√ß√£o entre x e y n√£o seja linear ‚Äî isso √© um sinal para investigar modelos mais complexos.

### üìâ Erro Quadr√°tico M√©dio (MSE)
O Erro Quadr√°tico M√©dio (do ingl√™s Mean Squared Error ‚Äì MSE) √© uma m√©trica fundamental para avaliar a qualidade do ajuste da reta de regress√£o. Ele mede, em m√©dia, o qu√£o distantes est√£o os valores previstos dos valores reais, penalizando erros maiores com mais intensidade.


### üßÆ C√°lculo Manual do MSE em Python

```python
# C√°lculo manual do Erro Quadr√°tico M√©dio (MSE)
soma_dos_erros_quadrados = np.sum((y - y_pred) ** 2)
mse = soma_dos_erros_quadrados / len(y)

print(f"Erro Quadr√°tico M√©dio (MSE): {mse:.4f}")
```

Quanto menor o valor de MSE, melhor a reta se ajusta aos dados.

Um MSE = 0 significaria que a reta passa exatamente por todos os pontos (o que √© raro em dados reais).
