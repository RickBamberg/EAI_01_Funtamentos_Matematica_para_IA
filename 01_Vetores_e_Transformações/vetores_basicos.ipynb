
# Vetores B√°sicos em 2D e 3D

Este notebook mostra como representar vetores em 2D e 3D, calcular magnitude, normalizar vetores, aplicar opera√ß√µes vetoriais e visualizar suas representa√ß√µes gr√°ficas.

---

## üß† Resumo Conceitual

- Um **vetor** representa dire√ß√£o, sentido e magnitude.
- Vetores 2D: `(x, y)`; Vetores 3D: `(x, y, z)`
- A **magnitude** √© o "tamanho" do vetor (dist√¢ncia da origem).
- Vetor **normalizado** tem magnitude igual a 1.
- Opera√ß√µes vetoriais: soma, subtra√ß√£o, produto escalar.
- A **visualiza√ß√£o gr√°fica** ajuda a entender a composi√ß√£o vetorial.

---


```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

class Vetor2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Vetor2D({self.x}, {self.y})"

    def __add__(self, outro):
        return Vetor2D(self.x + outro.x, self.y + outro.y)

    def __sub__(self, outro):
        return Vetor2D(self.x - outro.x, self.y - outro.y)

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2)

    def normalizado(self):
        mag = self.magnitude()
        return Vetor2D(self.x / mag, self.y / mag) if mag != 0 else Vetor2D(0, 0)

    def produto_escalar(self, outro):
        return self.x * outro.x + self.y * outro.y
```

```python
v1 = Vetor2D(3, 4)
v2 = Vetor2D(1, 2)
v_soma = v1 + v2

print("v1:", v1)
print("v2:", v2)
print("Soma (v1 + v2):", v_soma)
print("Produto escalar:", v1.produto_escalar(v2))
```

```python
def plotar_vetores_2d(vetores, cores, labels):
    plt.figure(figsize=(6, 6))
    for v, cor, label in zip(vetores, cores, labels):
        plt.quiver(0, 0, v.x, v.y, angles='xy', scale_units='xy', scale=1, color=cor, label=label)
    plt.xlim(-1, 6)
    plt.ylim(-1, 6)
    plt.axhline(0, color='gray', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.title("Vetores 2D e sua Soma")
    plt.show()

plotar_vetores_2d([v1, v2, v_soma], ['r', 'b', 'g'], ['v1', 'v2', 'v1+v2'])
```

```python
class Vetor3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __repr__(self):
        return f"Vetor3D({self.x}, {self.y}, {self.z})"

    def __add__(self, outro):
        return Vetor3D(self.x + outro.x, self.y + outro.y, self.z + outro.z)

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

    def normalizado(self):
        mag = self.magnitude()
        return Vetor3D(self.x / mag, self.y / mag, self.z / mag) if mag != 0 else Vetor3D(0, 0, 0)
```

```python
v3a = Vetor3D(1, 2, 1)
v3b = Vetor3D(2, 1, 2)
v3_soma = v3a + v3b

print("v3a:", v3a)
print("v3b:", v3b)
print("Soma (v3a + v3b):", v3_soma)
```

```python
vetores_3d = [Vetor3D(1, 0, 0), Vetor3D(0, 1, 0), Vetor3D(0, 0, 1), Vetor3D(1, 1, 1)]
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Definindo cores antes de usar
cores = ['r', 'g', 'b', 'y']  # Vermelho, Verde, Azul, Amarelo

for v, cor in zip(vetores_3d, cores):
    ax.quiver(0, 0, 0, v.x, v.y, v.z, color=cor)

ax.set_xlim([-1, 2])
ax.set_ylim([-1, 2])
ax.set_zlim([-1, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Vetores 3D - Base Can√¥nica")
plt.show()
```

```python
fig = plt.figure(figsize=(7, 6))
ax = fig.add_subplot(111, projection='3d')
cores = ['r', 'b', 'g']
labels = ['v3a', 'v3b', 'v3a+v3b']
vetores = [v3a, v3b, v3_soma]

for v, cor, label in zip(vetores, cores, labels):
    ax.quiver(0, 0, 0, v.x, v.y, v.z, color=cor, label=label)

ax.set_xlim([0, 4])
ax.set_ylim([0, 4])
ax.set_zlim([0, 4])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Vetores 3D e sua Soma")
ax.legend()
plt.show()
```
